```python
# загрузка датасета
df = pd.read_csv('fgl.csv')
```

```python
df
```

```python
# отделение целевой функции
y = df['type']
x = df.drop(labels='type', axis=1)
print(x)
print(y)
```

```python
# разделение на тестовую и обучающую выборку
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
lr_list = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
n_est_list = [i for i in range(30, 120, 10)]
max_d_list = [i for i in range(2,6)]
print(n_est_list)
print(max_d_list)

best_score = 0

for max_depth in max_d_list:
  for n_estimators in n_est_list:
    for learning_rate in lr_list:
      gb_clf = GradientBoostingClassifier(n_estimators=20, learning_rate=learning_rate, max_depth=max_depth)
      gb_clf.fit(X_train, y_train)

      print("Learning rate: ", learning_rate)
      print("Accuracy score (training): {0:.3f}".format(gb_clf.score(X_train, y_train)))
      print("Accuracy score (validation): {0:.3f}".format(gb_clf.score(X_test, y_test)))

      if gb_clf.score(X_test, y_test) > best_score:
        best_cls = gb_clf
    print('='*20)
```

```python
index = 1
print(y.iloc[[index]])
print(best_cls.predict(x.iloc[[index]]))
```
